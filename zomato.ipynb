### Droping Unwanted Columns

```python
remove_cols = ['URL','Timing', 'Full_Address','PhoneNumber', 'PeopleKnownFor','PopularDishes' ]
df = df.drop(columns=remove_cols)
```

### 1. Handling Null Values

```python
df.isna().sum()
```

```python
# In the column KnownFor has 90 percent null values so we can drop it
df = df.drop(columns='KnownFor')
```

```python
df.isna().sum()
```

### 2. Handling Inconsistent Formatting

```python
df.info()
```

```python
df['Dinner Ratings'].unique()
```

```python
series = df[df['Dinner Ratings'] != '-']['Dinner Ratings'].astype(float)
print(series.mean(),series.median())
```

```python
df['Dinner Ratings'] = df['Dinner Ratings'].replace('-',round(series.mean(),1))
```

```python
df['Dinner Ratings'].unique()
```

```python
df.info()
```

```python
df['Delivery Ratings'].unique()
```

```python
series = df[df['Delivery Ratings'] != '-']['Delivery Ratings'].astype(float)
print(series.mean(),series.median())
```

```python
df['Delivery Ratings'] = df['Delivery Ratings'].replace('-',round(series.mean(),1))
```

```python
df['Delivery Ratings'].unique()
```

```python
df.info()
```

### 3. Correcting Data types

```python
df.info()
```

```python
df['Dinner Ratings'] = df['Dinner Ratings'].astype(float)
df['Delivery Ratings'] = df['Delivery Ratings'].astype(float)
```
