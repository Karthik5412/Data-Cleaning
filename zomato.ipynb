### 5. Feature Extraction

```python
df.head()
```

```python
df.columns
```

```python
# Removing bangalore from Area column because we are working on bangalore dataset
def area_op(x):
    return x.split(',')[0]

df['Area'] = df['Area'].apply(area_op)
```

```python
df.head(2)
```

```python
# Creating a column called overall rating
df['Total Rating'] = (df['Delivery Ratings'] * 0.6) + (df['Delivery Ratings'] * 0.4)
```

```python
df.head(2)
```

```python
# Deriving a new coloumn by calculating average cost in a region
avg_area = df.groupby('Area')['AverageCost'].mean().to_dict()
```

```python
df['Area Cost'] = df['Area'].map(avg_area)
df['Area Cost'] = round(df['Area Cost'],0)
```

```python
# Creating a new column called discount
df['Discount'] = df['AverageCost'] - df['Area Cost']
df.loc[df['Discount'] < 0,'Discount'] = 0
```

```python
df[df['Discount'] <0]
```

```python
df['Cuisines'].unique().tolist()
```

```python
# Counting the dishes of hotel
def dishes_count(x):
    li = list(x.split(','))
    return len(li)

df['Total Items'] = df['Cuisines'].apply(dishes_count)
```

```python
df['Total Items'].value_counts()
```

### 6. Eleminating Outliers

```python
df.info()
```

```python
cols = ['Dinner Ratings','Delivery Ratings','Dinner Reviews','Delivery Reviews','AverageCost','Total Rating','Area Cost','Discount']
```

```python
for i in cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(df[i])
    
```

```python
import math
```

```python
# Dinner Rating
df['Dinner Ratings'].unique()
```

```python
q1 = np.percentile(df['Dinner Ratings'] ,25)
q3 = np.percentile(df['Dinner Ratings'] ,75)

iqr = q3 - q1

uf = q3 + 1.5 * iqr
lf = q1 - 1.5 * iqr

print(lf,uf)
```

```python
# Here we are unable to find outliers in this regular iqr method so we are using MAD method
x = df['Dinner Ratings']
mid = df['Dinner Ratings'].median()
mad_in = np.abs(x - mid)
mad = mad_in.median()
```

```python
lf = mid - (3 * mad *1.486)
uf = mid - (3 * mad *1.486)

print(lf,uf)
```

```python
print(lf,uf)
```

```python
# Delivery Ratings
q1 = np.percentile(df['Delivery Ratings'],25)
q3 = np.percentile(df['Delivery Ratings'],75)

iqr = q3 - q1

uf = q3 + 1.5 * iqr
lf = q1 - 1.5 * iqr

print(math.floor(uf * 10) / 10,math.ceil(lf * 10) /10)
```

```python
df.loc[df['Delivery Ratings'] > uf,'Delivery Ratings'] = math.floor(uf * 10) / 10
df.loc[df['Delivery Ratings'] < lf,'Delivery Ratings'] = math.ceil(lf * 10) /10
```

```python
# Dinner Reviews
q1 = np.percentile(df['Dinner Reviews'],25)
q3 = np.percentile(df['Dinner Reviews'],75)

iqr = q3 - q1

uf = q3 + 1.5 * iqr
lf = q1 - 1.5 * iqr

print(uf,lf)
```

```python
df.loc[df['Dinner Reviews'] > uf,'Dinner Reviews'] = uf.astype(int)
df.loc[df['Dinner Reviews'] < lf,'Dinner Reviews'] = lf.astype(int)
```

```python
# Delivery Reviews
q1 = np.percentile(df['Delivery Reviews'],25)
q3 = np.percentile(df['Delivery Reviews'],75)

iqr = q3 - q1

uf = q3 + 1.5 * iqr
lf = q1 - 1.5 * iqr

print(uf,lf)
```

```python
df.loc[df['Delivery Reviews'] > uf,'Delivery Reviews'] = uf.astype(int)
df.loc[df['Delivery Reviews'] < lf,'Delivery Reviews'] = lf.astype(int)
```

```python
# AverageCost
q1 = np.percentile(df['AverageCost'],25)
q3 = np.percentile(df['AverageCost'],75)

iqr = q3 - q1

uf = q3 + 1.5 * iqr
lf = q1 - 1.5 * iqr

print(uf,lf)
```

```python
df.loc[df['AverageCost'] > uf,'AverageCost'] = uf.astype(int)
df.loc[df['AverageCost'] < lf,'AverageCost'] = lf.astype(int)
```

```python
# Area Cost
q1 = np.percentile(df['Area Cost'],25)
q3 = np.percentile(df['Area Cost'],75)

iqr = q3 - q1

uf = q3 + 1.5 * iqr
lf = q1 - 1.5 * iqr

print(uf,lf)
```

```python
# Replacing this Dinner Ratings column with labels
def rating(x):
    if x <= 3.5:
        return 'Poor'
    elif x > 3.5 and x <= 4:
        return 'Average'
    elif x > 4 and x <= 4.5:
        return 'Good'
    else:
        return 'Best'
    
df['Dinner Ratings'] = df['Dinner Ratings'].apply(rating)
            
    
```
